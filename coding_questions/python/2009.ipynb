```python
import pandas as pd

# Task: find users who are a viever and also a streamer
# DataFrame: twitch_sessions

# solution: 
# in the session_type there were two distinct values: 'viewer' and 'streamer'
# after some EDA, it should be noted that the viewer and streamer mode should not be the same session, and just has the user(user_id) ever streamed alos viewed.
```

```python
twitch_sessions.groupby("user_id")["session_type"].nunique() \
    .reset_index(name="types") \
    .loc[lambda df: df["types"] > 1, "user_id"]
```
